Top 100 directors from https://www.independent.co.uk/arts-entertainment/films/news/best-directors-100-years-richard-linklater-a7566781.html

```python
directorNamesFile = "./data/top100directors.txt"
with open(directorNamesFile) as f:
    directorNames = [l.strip() for l in f.readlines()] 
len(directorNames)
```

```python
import pandas as pd
```

```python
imdbNameData = "./data/data.tsv"
df=pd.DataFrame()
for chunk in pd.read_csv(imdbNameData, sep='\t', chunksize = 100000):
    df = df.append(chunk[chunk.primaryName.isin(directorNames)][['nconst', 'primaryName']])
```

```python
results = df.drop_duplicates(subset='primaryName', keep="first")
```

```python
ids = results.nconst.values.tolist()
mapping = dict(results.values.tolist())
mapping
```
